```python
def mentioned(string):
    if type(string) == str:
        if ('elon' or 'musk' or 'tesla' or 'spacex' or 'boring co') in string.lower():
            return 1.0
        elif 'Boring' in string:
            return 1.0
        else:
            return 0.0
    else:
        return 0.0
```

```python
def get_polarity(string):
    if type(string) == str:
        return TextBlob(string).sentiment.polarity
def get_subjectivity(string):
    if type(string) == str:
        return TextBlob(string).sentiment.subjectivity
```
